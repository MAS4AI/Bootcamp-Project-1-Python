```python
def MAPs():
    '''
    Function to choose a map for the game.
    
    Returns:
        str: The selected map name.
    '''
    Maps={"1":'Al Mazrah',"2":'Al Malik International',"3":'Dome'}
    while True:
        
        print('1. Al Mazrah ')
        print('2. Al Malik International ')
        print('3. Dome')
        print('0. Exit')
        
        user_map = input('chose one of these maps \n')
        if user_map =="1" or user_map =="2" or user_map =="3":
            Map =Maps['{}'.format(user_map)] 
            print('you picked {}'.format(Map))
            return Map
        if user_map == '0':
            break 
        else:
            print('------invalid input------')
            

            
```

```python
def Gears():
    
    '''
    Function to choose weapons and armor for the game.
    
    returns  weapon_dps | chosen_weapon | armor_dur (armor_Durability) | chosen_armor | dps_after_buff
    '''
    weapons ={"1":'Sword',"2":'Bow',"3":'Axe'}
    dps ={"Sword":75,"Bow":110,"Axe":115}
    armor = {"1":'Elite Knight Set',"2":'Havels Set',"3":'Catarina Set'}
    armor_durability ={"Elite Knight Set":4200,"Havels Set":3640,"Catarina Set":5000}
    chosen_weapon=None
    weapon_dps=0
    dps_after_buff=0
    chosen_armor=None
    armor_dur=0
    DPS_buff = lambda x: x*1.6
    
    while True:
        print('\n')
        print('1. for picking a weapon')
        print('2. for picking an armor')
        print('0. Exit')
        
        user_gear=input('chose what you want to modife >')
        
        if user_gear == '1': 
            print('\n')
            print('1. Sword')
            print('2. Bow')
            print('3. Axe \n')
            user_weapon = input('pick your weapon >')
            
            if user_weapon =='1'or user_weapon =='2'or user_weapon =='3':
                chosen_weapon =weapons['{}'.format(user_weapon)]
                weapon_dps = dps['{}'.format(chosen_weapon)]
                dps_after_buff=DPS_buff(weapon_dps)
                print(weapon_dps)
            else:
                print('------invalid input------')
                
        elif user_gear == '2': 
            print('\n')
            print('1. Elite Knight Set')
            print('2. Havels Set')
            print('3. Catarina Set')  
            user_armor = input('pick your weapon >')
            
            if user_armor =='1'or user_armor =='2'or user_armor =='3':
                chosen_armor =armor['{}'.format(user_armor)]
                armor_dur = armor_durability['{}'.format(chosen_armor)]
                print(armor_dur)
            else:
                print('------invalid input------')
            ## need to be completed 
        elif user_gear =='0':
            break
        
        else:
             print('------invalid input------')
                
    return weapon_dps,chosen_weapon,armor_dur,chosen_armor,dps_after_buff
            
```

```python
def Friends():
    '''
    Function to invite a friend to play the game.

    Returns:
        str: The name of the invited friend.
    '''
    friends = {'majed':'online',
               'ahmed':'offline',
               'rawan':'online',
               'abear':'offline',}
    while True:
        print('Invite your friend \n')
        print(*friends.keys())
        print('0. Exit')
        user_friend = input('chose your friend name > ')
        if user_friend=='0':
            break
        try:
            if friends[user_friend] == 'offline':
                print('this player offline \n')
            elif friends[user_friend] == 'online':
                print('invation has been sent ...')
                player_name = user_friend
                return player_name
        except:
            print('------invalid input------')
    
```

```python
def Play_gmae(map_name,chosen_weapon,weapon_dps,chosen_armor,armor_dur,player_name,friend_name,dps_after_buff):
    """
    Function to play the game with selected options.

    Args:
        map_name (str): The name of the selected map.
        chosen_weapon (str): The chosen weapon.
        weapon_dps (int): The DPS of the chosen weapon.
        chosen_armor (str): The chosen armor.
        armor_dur (int): The durability of the chosen armor.
        player_name (str): The player's name.
        friend_name (str): The friend's name.
        dps_after_buff (float): The DPS of the weapon after applying the buff.
        
        """
    
    print('player name: {}'.format(player_name))
    print('Map is: {}'.format(map_name))
    print('your weapon is: {}| Weapon DPS {} | dps with buff {}'.format(chosen_weapon,weapon_dps,dps_after_buff))
    print('your armor is: {} |armor durability {}'.format(chosen_armor,armor_dur))
    print('your friend player is : {}'.format(friend_name))
    print('good luck in your fight!!!')
    
    
```

```python
def main(): 
    """
    Main function to run the game.

    Returns:
        None
    """
    Map_User = None
    
    weapon_name=None
    armor_name=None
    weapon = 0
    dps_after_buff=0
    armor_Durability = 0

    friend_name =None
    player_name =input('enter your name: ')
    while True:

        print('1. Chose the map')
        print('2. Modify your gears')
        print('3. Invite a friend')
        print('4. Play the game')
        print('0. Exit')
        
        user_input = input('enter you Choice > ')
        
        if user_input == '1':
            Map_User = MAPs()
        elif user_input =='2':
            weapon,weapon_name, armor_Durability,armor_name,dps_after_buff =Gears()
            print(weapon,weapon_name,armor_Durability,armor_name,' -----------------')
        elif user_input =='3':
            friend_name= Friends()
            print('name>',friend_name)
        elif user_input =="4":
            Play_gmae(Map_User,weapon_name,weapon,armor_name,armor_Durability,player_name,friend_name,dps_after_buff)
            break
        elif user_input == '0':
            break
        else:
            print('------invalid input------')
```

```python
main()
```
